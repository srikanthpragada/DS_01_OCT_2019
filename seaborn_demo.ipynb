```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
```

### Categorical plots

```python
sns.boxplot(x="day",y="tip", data = tips)
```

```python
sns.boxplot(x="sex",y="total_bill",data = tips)
```

```python
sns.boxplot(x="day",y="tip", hue="time", data = tips)
```

```python
sns.barplot(x="day",y="total_bill", hue="sex", data = tips, ci=False)
```

```python
sns.catplot(x="day",y="total_bill",data=tips, hue="sex",col="smoker", kind='box')
```

```python
seltips = tips[ tips['size'] <= 4]
sns.catplot(x="day",y="total_bill",data=seltips,col="size", kind='box')
plt.gcf().set_size_inches(10,10)  # Change current figure size 
```

```python
sns.countplot(x="day",data=tips)
```

```python
flights = sns.load_dataset("flights")`
```

```python
flights.head()
```

```python
flights.groupby('year')['passengers'].sum()
```

```python
sns.barplot(x="year",y="passengers",data=flights,ci=False,estimator=np.sum)
```

### Relational plots

```python
sns.scatterplot(x="total_bill",y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex", data=tips)
```

```python
fs = flights.groupby('year',as_index=False)['passengers'].sum()
fs
```

```python
sns.lineplot()
```

```python
sns.lineplot(x='year',y='passengers',data=fs)
plt.xlim((1945,1965))
```

```python
sns.relplot(x="total_bill",y="tip",col="day", hue='sex',data=tips)
```

```python
sns.lineplot(x="year",y="passengers",data=flights, estimator=np.sum)
```

## pairplot

```python
sns.pairplot(tips, hue='sex',vars=['total_bill','tip'], diag_kind='hist')
plt.gcf().set_size_inches(8,8)
```

```python
sns.pairplot(tips)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips)
```

```python
sns.rugplot(tips['tip'])
```

### Regression Plot

```python
sns.regplot(x="total_bill",y='tip', data=tips)
```

```python
sns.lmplot(x="total_bill",y='tip',hue="smoker", data=tips,ci=False)
```

```python
sns.lmplot(x="total_bill",y='tip',hue="time", col='day', data=tips,ci=False)
```

```python
sns.residplot(x="total_bill",y='tip', data=tips)
```
